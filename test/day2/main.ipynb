1. D liu ri rc (Discrete data):
- D liu c c t vic **m** (s nguyn)
- m ch s lng (ch th l s nguyn ri)

- V d:
+ S hc sinh trong lp (40, 100, 200)
+ S con mo (3, 4, 2)
+ S li chnh t trong bi vit (5, 10, 15)
+ S xe trong bi  (10, 100, 45)
+ S ln truy cp website (1, 2, 3)
+ Cht lng sn phm (tt: bao nhiu ? kh: bao nhiu ? trung bnh: bao nhiu ?)
+ Trong R: d liu ri rc c biu din bng kiu integer hay factor (phn loi)
+ Biu  ph hp: bar chart (ct) - chiu cao ca cc ct th hin r rng cc gi tr khng lin tc (mang tnh phn loi)
+ V d: so snh doanh s bn hng gia cc thng (Thng 1, 2, 3), so snh s lng sinh vin ca tng lp

+ Biu  thanh ngang (Horizontal Bar Chart): tng t biu  ct nhng cc thanh nm ngang hu ch khi tn biu din cho ct qu di nn quay ngang nhn cho d
+ V d: danh sch cc quc gia vi dn s tng ng


2. D liu lin tc (Continuous data):
- D liu c c t vic **o lng** (thng l s thc)
- V d:
+ Chiu cao (170.5cm, 150cm, 165cm, 176cm)
+ Cn nng (60.2kg, 56.7kg, 80kg, 75.4kg)
+ Thi gian (2.35s, 5.6s, 7.8s, 2.3s)
+ Nhit  (36.7  C, 67.8  C)

+ Trong R: d liu lin tc c biu din bi kiu d liu numeric (s thc)
+ Biu  ph hp:
+ p dng cho 1 bin (1 ct - 1 vector): Histogram (xem phn phi, tn sut ca d liu), Boxplot (xem median, t phn v, so snh nhm, pht hin outliers), Density plot (ging histogram nhng mt hn)
+ p dng cho 2 bin tr ln (2 ct - 2 vector): scatter plot (xem mi quan h gia 2 bin lin tc), line chart (theo di xu hng thi gian), heatmap (xem tng quan gia nhiu bin cng lc)
