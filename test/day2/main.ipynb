1. Dữ liệu rời rạc (Discrete data):
- Dữ liệu có được từ việc **đếm** (số nguyên)
- Ám chỉ số lượng (địch thị là số nguyên rồi)

- Ví dụ:
+ Số học sinh trong lớp (40, 100, 200)
+ Số con mèo (3, 4, 2)
+ Số lỗi chính tả trong bài viết (5, 10, 15)
+ Số xe trong bãi đỗ (10, 100, 45)
+ Số lần truy cập website (1, 2, 3)
+ Chất lượng sản phẩm (tốt: bao nhiêu ? khá: bao nhiêu ? trung bình: bao nhiêu ?)
+ Trong R: dữ liệu rời rạc được biểu diễn bằng kiểu integer hay factor (phân loại)
+ Biểu đồ phù hợp: bar chart (cột) - chiều cao của các cột thể hiện rõ ràng các giá trị không liên tục (mang tính phân loại)
+ Ví dụ: so sánh doanh số bán hàng giữa các tháng (Tháng 1, 2, 3), so sánh số lượng sinh viên của từng lớp

+ Biểu đồ thanh ngang (Horizontal Bar Chart): tương tự biểu đồ cột nhưng các thanh nằm ngang hữu ích khi tên biểu diễn cho cột quá dài nên quay ngang nhìn cho dễ
+ Ví dụ: danh sách các quốc gia với dân số tương ứng


2. Dữ liệu liên tục (Continuous data):
- Dữ liệu có được từ việc **đo lường** (thường là số thực)
- Ví dụ:
+ Chiều cao (170.5cm, 150cm, 165cm, 176cm)
+ Cân nặng (60.2kg, 56.7kg, 80kg, 75.4kg)
+ Thời gian (2.35s, 5.6s, 7.8s, 2.3s)
+ Nhiệt độ (36.7 độ C, 67.8 độ C)

+ Trong R: dữ liệu liên tục được biểu diễn bởi kiểu dữ liệu numeric (số thực)
+ Biểu đồ phù hợp:
+ Áp dụng cho 1 biến (1 cột - 1 vector): Histogram (xem phân phối, tần suất của dữ liệu), Boxplot (xem median, tứ phân vị, so sánh nhóm, phát hiện outliers), Density plot (giống histogram nhưng mượt hơn)
+ Áp dụng cho 2 biến trở lên (2 cột - 2 vector): scatter plot (xem mối quan hệ giữa 2 biến liên tục), line chart (theo dõi xu hướng thời gian), heatmap (xem tương quan giữa nhiều biến cùng lúc)

1. Biểu đồ cột barplot trong thống kê
+ Một trong những loại biểu đồ phổ biến trực quan nhất, dùng để so sánh giá trị giữa các nhóm/danh mục (category - nhãn (label) dùng để phân loại dữ liệu)
+ Ví dụ: ta có 2 giới tính "Nam", "Nữ" là các danh mục, hay tốt, khá, trung bình mỗi mức là 1 danh mục, thậm chí là các màu đỏ xanh vàng cũng là biến phân loại các danh mục
+ Barplot phù hợp trong trường hợp: so sanh số lượng, tần suất hay giá trị trung bình giữa các nhóm
+ Biểu diễn dữ liệu định tính (categorical) kết hợp với 1 biến định lượng (numeric)
+ Ví dụ:
+ So sánh doanh thu theo tháng, số lượng sản phẩm, điểm trung bình theo lớp học

```r
# So sánh số lần xuất hiện giữa các nhóm
counts <- c(40, 25, 15)
names(counts) <- c("A", "B", "C")

barplot(counts, xlab = "Nhóm", ylab = "Số lần xuất hiện", main = "Simple Barplot", col = "skyblue")
```

```r
# Khi tên danh mục dài dùng (Horizontal Barplot)
barplot(
    sort(counts),
    horiz = TRUE,
    col = "#dbc59b",
    main = "Horizontal Barplot"
)
```

```r
# Grouped Barplot (Barplot theo nhóm)
data <- matrix(c(9,7,8,6), nrow = 2)

colnames(data) <- c("An", "Bình")
rownames(data) <- c("Toán", "Văn")

data
```

```r
# So sánh điểm toán vs văn giữa các học sinh
# beside = TRUE các cột đứng cạnh nhau
barplot(
    data,
    beside = TRUE,
    col = c("red", "blue"),
    legend = TRUE
)
```

```r
# Stacked Barplot
barplot(data, beside = FALSE, col = c("red", "blue"), legend = TRUE)
```

```r
# 100% Barplot
prop_data <- prop.table(data, 2)
prop_data
```

```r
barplot(prop_data, col = c("red", "blue"), legend = TRUE)
```

```r
barplot(sort(counts, decreasing = TRUE), col = "green")
```

```r
library(ggplot2)

df <- data.frame(
    Value = c(40,25,15),
    Group = c("A","B","C")
)

ggplot(df, aes(x=reorder(Group,-Value), y=Value)) + geom_bar(stat = "identity", fill = "steelblue") + theme_minimal()
```

```r
# MINI PROJECT: Phân tích doanh số sản phẩm
df <- data.frame(
    Product = c("A", "B", "C", "D"),
    Q1 = c(120, 90, 60, 150),
    Q2 = c(100, 110, 80, 130)
)

df
```

```r
df$Total <- df$Q1 + df$Q2
```

```r
df
```

```r
# Vẽ 4 loại Barplot
# So sánh tổng doanh số
barplot(
    df$Total,
    xlab = "Category",
    ylab = "Total",
    col = "lightblue", # skyblue
    main = "Total sales by Product",
    names.arg = df$Product
)
```

```r
# Tìm sản phẩm bán chạy nhất
# barplot(
#     sort(df$Total, decreasing = TRUE),
#     xlab = "Category",
#     ylab = "Total",
#     col = "lightblue", # skyblue
#     main = "Total sales by Product",
#     names.arg = df$Product
# )

totals <- setNames(df$Total, df$Product)
barplot(
    sort(totals, decreasing = TRUE),
    xlab = "Category",
    ylab = "Total",
    col = "orange", # skyblue
    main = "Total sales (Descending)",
)
```

```r
# 3. Grouped Barplot - so sánh Q1 VS Q2
print(df)
sales_matrix <- rbind(df$Q1, df$Q2)
sales_matrix
```

```r
colnames(sales_matrix) <- df$Product
sales_matrix
```

```r
rownames(sales_matrix) <- c("Q1", "Q2")
sales_matrix
```

```r
# So sánh tăng giảm giữa hai quý
barplot(
    sales_matrix,
    beside = TRUE,
    col = c("red", "blue"),
    main = "Sales Comparison by Q1 vs Q2",
    legend = TRUE
)
```

```r
# Xem tỷ trọng Q1 và Q2 trong tổng
barplot(
    sales_matrix,
    beside = FALSE,
    legend = TRUE,
    col = c("red", "blue"),
    main = "Sales Structure"
)
```

```r
library(ggplot2)
library(tidyr)

df_long <- pivot_longer(
    df,
    cols = c(Q1, Q2),
    names_to = "Quarter",
    values_to = "Sales"
)

df_long
```

```r
ggplot(df_long, aes(x = Product, y = Sales, fill = Quarter)) +
    geom_bar(stat = "identity") +
    labs(title = "Sales Structure",
        x = "Product",
        y = "Sales") +
    theme_minimal()
```

```r
ggplot(df_long, aes(x = Product, y = Sales, fill = Quarter)) +
    geom_bar(stat = "identity", position = "fill") +
    labs(title = "Sales Structure (%)",
        y = "Proportion") +
    theme_minimal()
```
